```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import LabelEncoder
```

```python
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
```

```python
features = ['east_rank_number', 
            'west_rank_number', 
            'east_winstreak',
            'west_winstreak', 
            'east_ratio', 
            'west_ratio', 
            'east_wins', 
            'west_wins', 
            'east_weight', 
            'east_height', 
            'west_weight',
            'west_height', 
            'east_n1_win_move', 
            'east_n1_win_percent',
            'east_n1_loss_move', 
            'east_n1_loss_percent', 
            'west_n1_win_move',
            'west_n1_win_percent', 
            'west_n1_loss_move', 
            'west_n1_loss_percent',
            'east_n2_win_move', 
            'east_n2_win_percent', 
            'east_n2_loss_move',
            'east_n2_loss_percent', 
            'west_n2_win_move', 
            'west_n2_win_percent',
            'west_n2_loss_move', 
            'west_n2_loss_percent', 
            'east_n3_win_move',
            'east_n3_win_percent', 
            'east_n3_loss_move', 
            'east_n3_loss_percent',
            'west_n3_win_move', 
            'west_n3_win_percent', 
            'west_n3_loss_move',
            'west_n3_loss_percent']

ml_features = df_matches[features]

ml_features
```

```python
ml_target = df_matches['winner']

ml_target
```

```python
ml_features_train, ml_features_test, ml_target_train, ml_target_test = train_test_split(ml_features, ml_target, test_size=0.2, random_state=42)

clf = KNeighborsClassifier()
clf.fit(ml_features_train, ml_target_train)

model_test = clf.predict(ml_features_test)
```

```python
cv_scores = cross_val_score(clf, ml_features, ml_target, cv=5)

print(f"Cross-validation scores: {cv_scores}")
print(f"Mean cross-validation score: {cv_scores.mean():.4f}")
```

```python
accuracy = accuracy_score(ml_target_test, model_test)
precision = precision_score(ml_target_test, model_test, average='weighted')
recall = recall_score(ml_target_test, model_test, average='weighted')
f1 = f1_score(ml_target_test, model_test, average='weighted')

print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"Precision: {precision * 100:.2f}%")
print(f"Recall: {recall * 100:.2f}%")
print(f"F1-score: {f1 * 100:.2f}%")
```
